```python
df_stock_list=pd.read_excel("D:/Stock_monitoring_list1.xlsx")
```

```python
L_stock_code=df_stock_list["Stock code"].tolist()
```

```python
len(L_stock_code)
```

```python
def return_max1(from_date,L_stock_code,number_max=5):
    L1=[]
    for i in L_stock_code:
        message="Now downloading data for "+i+":"
        print(message)
        df=yf.download(i,from_date,progress=False)
        return1=100*(df["Close"].iloc[-1]-df["Close"].iloc[0])/df["Close"].iloc[0]
        L1.append(return1)
    df_1=pd.DataFrame({"Stock_code":L_stock_code,
                 "Return":L1})
    df2=df_1.sort_values(["Return"],ascending=False).iloc[:number_max,]
    L_stock_top5=df2["Stock_code"].tolist()
    return(L_stock_top5)
```

```python
L_stocks_top5=return_max1("2019-09-01",L_stock_code)
L_stocks_top5
```

```python
def return_max2(from_date,L_stock_code,number_max=5):
    L1=[]
    D1={}
    for i in L_stock_code:
        message="Now downloading data for "+i+":"
        print(message)
        df=yf.download(i,from_date,progress=False)
        return1=100*(df["Close"].iloc[-1]-df["Close"].iloc[0])/df["Close"].iloc[0]
        L1.append(return1)
    df_1=pd.DataFrame({"Stock_code":L_stock_code,
                 "Return":L1})
    df2=df_1.sort_values(["Return"],ascending=False).iloc[:number_max,]
    L_stock_top5=df2["Stock_code"].tolist()
    L_return_top5=df2["Return"].tolist()
    for i in range(len(L_stock_top5)):
        D1[L_stock_top5[i]]=L_return_top5[i]
    return(D1)
```

```python
D_stocks_top5=return_max2("2019-09-01",L_stock_code)
D_stocks_top5
```
