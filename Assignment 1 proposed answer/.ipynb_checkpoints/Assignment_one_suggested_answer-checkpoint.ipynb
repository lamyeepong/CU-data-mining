## Question two

```python
df=yf.download("^HSI","2018-01-01","2020-12-14")
```

```python
df["Voltatity"]=df["High"]-df["Low"]
```

```python
df["GAP"]=df["Open"]-df["Close"].shift()
```

```python
df["SMA50"]=df["Close"].rolling(window=50).mean()
```

```python
df.tail()
```

```python
df["Date"]=df.index
df["Year"]=df["Date"].dt.year
df["Month"]=df["Date"].dt.month
```

```python
df.tail()
```

```python
df_monthly=df.groupby(["Year","Month"]).agg({'Close':['max','min']})
```

```python
df_monthly["diff"]=df_monthly["Close"]["max"]-df_monthly["Close"]["min"]
```
