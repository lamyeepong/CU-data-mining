```python
import pandas as pd
from datetime import date
import yfinance as yf
```

# Question One

```python
def read_stock_xlsx():
    df = pd.read_excel("Stock_monitoring_list1.xlsx")
    return list(df["Stock code"])

def get_return(index_list, input_date):
    today = date.today()
    d1 = today.strftime("%Y-%m-%d")
    re_list = []
    for i in index_list:
        df1=yf.download(i,input_date,d1)
        re = 100*(df1["Close"][-1] - df1["Close"][0])/df1["Close"][0]
        re_list.append(re)
    return re_list

def run():
    year = input("Please input start year")
    month = input("Please input start month")
    day = input("Please input start day")
    input_date = year + "-" + month + "-" + day
    index_list = read_stock_xlsx()
    df = pd.DataFrame({"Stock code":index_list, "Return":get_return(index_list,input_date)})
    return [df["Stock code"][i] for i in list(df["Return"].sort_values(ascending=False).iloc[:5].index)]
```

```python
run()
```

# Question 2

```python
df2 = yf.download("^HSI", "2018-01-01", "2020-12-14")
```

```python
df2.head()
```

```python
df2["Volatility"] = df2["High"] - df2["Low"]
df2.head()
```

```python
df2["GAP"] = df2["Open"] - df2["Close"].shift(periods=1)
df2.head()
```

```python
df2.tail()
```

```python
df2["SMA50"] = df2["Close"].rolling(window = 50).mean()
df2.tail()
```

```python
df2["Month"] = pd.to_datetime(df2.index, "%Y-%m-%d")
df2["Year"] =  df2["Month"].dt.year 
df2["Month"] = df2["Year"].apply(lambda x: str(x)) + "-" + df2["Month"].dt.month.apply(lambda x: str(x).zfill(2))
df3 = df2[["Month","Close"]].groupby("Month").max()
df3.rename(columns={"Close": "Close_Max"}, inplace= True)
df4 = df2[["Month","Close"]].groupby("Month").min()
df4.rename(columns={"Close": "Close_Min"}, inplace= True)
df5=pd.concat([df3,df4], axis=1)
df5.head()
```
